```python
%matplotlib inline
import math
import time
import numpy as np
from myPkg import myPlot as d2l 
```

```python
def normal(x, mu, sigma):
    p = 1 / math.sqrt(2 * math.pi * sigma**2)
    return p * np.exp(-0.5 / sigma**2 * (x - mu)**2)

```

```python
# 再次使用numpy进行可视化
x = np.arange(-7, 7, 0.01)

# 均值和标准差对
params = [(0, 1), (0, 2), (3, 1)]
d2l.plot(x, [normal(x, mu, sigma) for mu, sigma in params],
         xlabel='x',
         ylabel='p(x)', figsize=(4.5, 2.5),
         legend=[f'mean {mu}, std {sigma}' for mu, sigma in params])
```

```python
def sigmoid(x):
    p = 1/(1+np.exp(-x))
    return p 

def err(x):
    return np.random.randn(1,x.size)*0.1
```

```python
# 再次使用numpy进行可视化
x = np.arange(0, 7, 0.05)
params = range(1,5);

d2l.plot_errbar(x, [i*sigmoid(x) for i in params], 
         [err(x) for i in params],
         xlabel='Times(s)',
         ylabel='p(x)', figsize=(4.5, 2.5),
         legend=[f'mul {mu}' for mu in params])
```
